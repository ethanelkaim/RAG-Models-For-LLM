```python
import numpy as np
from sentence_transformers import SentenceTransformer
import faiss

model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
def load_and_embedd_dataset(
        dataset_name: str = 'ErikCikalleshi/new_york_times_news_2000_2007',
        split: str = 'train',
        model: SentenceTransformer = SentenceTransformer('paraphrase-MiniLM-L6-v2'),
        text_field: str = 'content',
        rec_num: int = 400
) -> tuple:
    """
    Load a dataset and embedd the text field using a sentence-transformer model
    Args:
        dataset_name: The name of the dataset to load
        split: The split of the dataset to load
        model: The model to use for embedding
        text_field: The field in the dataset that contains the text
        rec_num: The number of records to load and embedd
    Returns:
        tuple: A tuple containing the dataset and the embeddings
    """
    from datasets import load_dataset

    print("Loading and embedding the dataset")

    dataset = load_dataset(dataset_name, split=split)
    embeddings = model.encode(dataset[text_field][:rec_num])

    print("Done!")
    return dataset, embeddings
```

```python
DATASET_NAME = 'ErikCikalleshi/new_york_times_news_2000_2007'

dataset, embeddings = load_and_embedd_dataset(
    dataset_name=DATASET_NAME,
    rec_num=400,
    model=model,
)
shape = embeddings.shape
```

```python
print(f"The embeddings shape: {embeddings.shape}")
```

```python
# Function to create and train a FAISS index with ANN
# nlist is the number of clusters
def create_faiss_index(embeddings: np.ndarray, nlist: int = 100):
    """
    Create and train a FAISS index from embeddings using Approximate Nearest Neighbors (ANN).

    Args:
        embeddings: The embeddings to index.
        nlist: The number of clusters to create for the ANN search.

    Returns:
        faiss.IndexIVFFlat: A trained FAISS index.
    """
    dimension = embeddings.shape[1]
    quantizer = faiss.IndexFlatL2(dimension)
    index = faiss.IndexIVFFlat(quantizer, dimension, nlist, faiss.METRIC_L2)

    index.train(embeddings.astype(np.float32))
    index.add(embeddings.astype(np.float32))
    return index
```

```python
faiss_index = create_faiss_index(embeddings)
```

```python
def augment_prompt(
        query: str,
        model: SentenceTransformer = SentenceTransformer('paraphrase-MiniLM-L6-v2'),
        index=None,
        dataset: dict = None,
        text_field: str = 'content',
) -> str:
    """
    Augment the prompt with the top 3 results from the knowledge base
    Args:
        query: The query to augment
        index: The FAISS index object
        dataset: The dataset containing the metadata
    Returns:
        str: The augmented prompt
    """
    query_embedding = model.encode(query).astype('float32').reshape(1, -1)
    distances, indices = index.search(query_embedding, k=3)
    text_matches = [dataset[text_field][i] for i in indices[0]]

    source_knowledge = "\n\n".join(text_matches)

    augmented_prompt = f"""Using the contexts below, answer the query.
    Contexts:
    {source_knowledge}
    If the answer is not included in the source knowledge - say that you don't know.
    Query: {query}"""

    return augmented_prompt, source_knowledge
```

```python
from langchain_experimental.text_splitter import SemanticChunker
import os
import numpy as np
from sentence_transformers import SentenceTransformer
import faiss
from datasets import load_dataset
from langchain_experimental.text_splitter import SemanticChunker
from langchain_openai.embeddings import OpenAIEmbeddings
```

```python

# Initialize your model for embeddings
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

```python
def load_and_chunk_dataset(
        dataset_name: str = 'ErikCikalleshi/new_york_times_news_2000_2007',
        split: str = 'train',
        model: SentenceTransformer = SentenceTransformer('paraphrase-MiniLM-L6-v2'),
        text_field: str = 'content',
        rec_num: int = 400,
        breakpoint_threshold_type='percentile',
        breakpoint_threshold_amount=90
) -> tuple:
    """
    Load a dataset and chunk the text field using semantic chunking with embeddings.
    """
    print("Loading and chunking the dataset")

    dataset = load_dataset(dataset_name, split=split)

    text_data = dataset[text_field][:rec_num]

    """
    embedding_model = SentenceTransformer('paraphrase-MiniLM-L6-v2')  # You may use SentenceTransformer for consistency
    text_splitter = SemanticChunker(
        embedding_model,
        breakpoint_threshold_type=breakpoint_threshold_type,
        breakpoint_threshold_amount=breakpoint_threshold_amount
    )
"""
    embedding_model = HuggingFaceEmbeddings(model_name='paraphrase-MiniLM-L6-v2')
    text_splitter = SemanticChunker(
        embedding_model,
        breakpoint_threshold_type=breakpoint_threshold_type,
        breakpoint_threshold_amount=breakpoint_threshold_amount
    )

    chunks = []
    for content in text_data:
        doc_chunks = text_splitter.create_documents([content])
        chunks.extend(doc_chunks)

    chunk_texts = [chunk.page_content for chunk in chunks]
    embeddings = model.encode(chunk_texts)

    print("Chunking and embedding completed!")
    return chunks, embeddings
```

```python

DATASET_NAME = 'ErikCikalleshi/new_york_times_news_2000_2007'
chunks, embeddings = load_and_chunk_dataset(
    dataset_name=DATASET_NAME,
    rec_num=400,
    model=model,
)
```

```python
# Create and train FAISS index with chunk embeddings
def create_faiss_index(embeddings: np.ndarray, nlist: int = 100):
    """
    Create and train a FAISS index from embeddings using Approximate Nearest Neighbors (ANN).
    """
    dimension = embeddings.shape[1]
    quantizer = faiss.IndexFlatL2(dimension)
    index = faiss.IndexIVFFlat(quantizer, dimension, nlist, faiss.METRIC_L2)

    index.train(embeddings.astype(np.float32))
    index.add(embeddings.astype(np.float32))
    return index

faiss_index = create_faiss_index(embeddings)
```

```python
def augment_prompt(
        query: str,
        model: SentenceTransformer = SentenceTransformer('paraphrase-MiniLM-L6-v2'),
        index=None,
        chunks=None,
) -> str:
    """
    Augment the prompt with the top 3 results from the knowledge base.
    """
    query_embedding = model.encode(query).astype('float32').reshape(1, -1)
    distances, indices = index.search(query_embedding, k=3)
    text_matches = [chunks[i].page_content for i in indices[0]]
    source_knowledge = "\n\n".join(text_matches)

    augmented_prompt = f"""Using the contexts below, answer the query.
    Contexts:
    {source_knowledge}
    If the answer is not included in the source knowledge - say that you don't know.
    Query: {query}"""

    return augmented_prompt, source_knowledge
```

```python
query = "How many jobs were added on Long Island in 2001?"
augmented_prompt, source_knowledge = augment_prompt(query, model=model, index=faiss_index, chunks=chunks)

import cohere
with open("chohere_api_keys.txt") as f:
    COHERE_API_KEY = f.read().strip()

co = cohere.Client(api_key=COHERE_API_KEY)
response = co.chat(
    model='command-r-plus',
    message=augmented_prompt,
)
print(response.text)
```
